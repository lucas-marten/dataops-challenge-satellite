```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import rasterio
```

```python
file = '/home/lucasmarten/dataops-challenge-satellite/dataops-challenge-satellite/satellite-data-processing/processed_data/tiffs/file1.epsg:4326.tiff'

dataset = xr.open_dataset(file)
data = dataset['band_data'].values
lons, lats = dataset.x.values, dataset.y.values
```

```python
fig = plt.figure(dpi=300)
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.contourf(lons, lats, data[0], cmap='gray')
plt.show()
```

```python
# Clip extreme values (2nd to 98th percentile)
p2, p98 = np.percentile(data, (2, 98))
band_06_stretched = np.clip(data, p2, p98)

# Normalize to 0-1 range
band_06_norm = (band_06_stretched - p2) / (p98 - p2)

fig = plt.figure(dpi=300)
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.contourf(lons, lats, band_06_norm[0], cmap='gray')
plt.show()
```
